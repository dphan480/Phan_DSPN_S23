Magic: the Gathering is a card game where players build decks representing spells, creatures, and lands in order to defeat their opponents. It is played between two players, who each start with 20 life. The goal is to bring their opponent's life down to 0.

The game consists of turns alternating between players. During their turn, players can do things like draw cards, play cards, and use cards on the battlefield.

There are many different types of cards, but the most important one is land. Why? Because lands provide the resources necessary to play your other cards! However, they tend not to do too much on their own.

This creates an interesting problem. You can only play cards that you have drawn, and aside from the seven that you start with, you only draw one card each turn. If you draw too few lands, you can't play your relevant cards. However, if you draw too many lands, you simply won't have relevant cards. These are respectively known as mana screwing and mana flooding.

A fundamental deckbuilding question arises: how many lands do you put in your deck? Too many and you increase your risk of mana flooding. Too few, mana screw. What's the optimal number? What'll make Goldilocks happy?

Companion
these are a special type of card. If you meet certain deckbuilding requirements (for example, Lurrus asks that your deck only contains cheap cards), you can begin the game with a companion card in your companion zone. While it's in there, it cannot be used in any way. However, during any one of your turns, you can pay 3 mana to put your companion into your hand. After that, it acts as a normal card in your hand; you still have to pay its mana value in order to play it.

Some of them are quite popular due to their low opportunity cost. For example, many decks already contain only cheap cards. In that case, it wouldn't cost much at all to have Lurrus as your companion. However, now these decks are guaranteed to have an expensive card they can play.


ARCHETYPE
Magic: the Gathering has many cards, and as such there are many potential decks you can create. As it turns out, these decks (at least the ones built to win) can be categorized into various gameplay strategies, also known as archetypes. The two I will be looking at are aggro decks and control decks.

Aggro (short for aggressive) decks want to bring their opponent's life down to 0 as soon as possible. They tend to do this by playing many cheap creatures, and as such do not need many lands to play all their cards. Since they want to end the game quickly, they are not planning to draw many cards, which makes the cards they do draw matter all the more. While other decks can handle drawing a few extra lands, for an aggro deck, drawing just a single extra land more than what they need can cause them to run out of steam and lose.

Control decks want to survive as long as possible. They do this by playing spells that let them stop their opponent's cards (aka control them). At the same time, they also try to accumulate resources, typically by drawing cards, so that as time goes on, they have more and more tools while their opponent has less and less. They plan on drawing many cards and having the resources to play these cards, so they tend to be fine with drawing a few extra lands. However, drawing just a single land less than what they need can stop them from playing a crucial kill spell before an aggro deck overruns them.


What is an mdfc?
MDFC stands for "modal dual-faced card." In this context, these are cards that you can play as either a land or a spell. To compensate for this flexibility, the land and spell are both worse than usual, so it's not a great idea to count these cards as full lands.


What is ramp?
Normally, you can only play one land a turn. So a player will typically have one land on turn one, two lands on turn two, etc. However, some cards let you accumulate resources faster than that. These are known as ramp cards. Getting more resources faster is great, but ramp cards are usually fragile or costly in some way, so only some decks play them. In additiona, drawing too many of these is very similar to drawing too many lands.


What is a cantrip?
A cheap card that draws a card. Many people specifically play cantrips to ensure that they can get (or not get) lands.


What is mana value?
It's basically the number of resources you need to play the card from your hand.



DATA COLLECTION:
Data was collected by scraping search results from mtgtop8.com, a website that collects tournament data.

Only decklists submitted for "major" or "professional" events were collected, in order to ensure that decks were built with winning in mind.

Data was gathered from the following formats: Standard, Pioneer, Alchemy, Historic, and Explorer. These formats were chosen because they are 1) popular, 2) share similar rulesets, and 3) do not contain many cards that break fundamental assumptions (mana value reflects actual cost, lands add one mana, etc.)

Only decklists from March 2021 onward were collected. This was chosen because cards that did break fundamental assumptions were banned from the aforementioned formats the month before.
